```python
import pandas

def read_dataset(prefix):
    """
    Read in all of the training and test data having the given prefix.
    """
    
    dfs = {}
    for name, suf, paired in [('u', '_unp.csv', False), ('d', '_disc.csv', True), ('c', '_conc.csv', True), ('b', '_bad_end.csv', False)]:
        fn = prefix + suf
        if os.path.exists(fn):
            dfs[name] = pandas.io.parsers.read_csv(fn, quoting=2)
        elif os.path.exists(fn + '.gz'):
            dfs[name] = pandas.io.parsers.read_csv(fn + '.gz', quoting=2, compression='gzip')
        elif os.path.exists(fn + '.bz2'):
            dfs[name] = pandas.io.parsers.read_csv(fn + '.bz2', quoting=2, compression='bz2')
        else:
            raise RuntimeError('No such file: "%s"' % fn)

    for df in dfs.itervalues():
        if df['correct'].count() == len(df['correct']):
            df['correct'] = df['correct'].map(lambda x: 1 if x == 'T' else 0)

    for df in [dfs['u'], dfs['b']]:
        diffv = df.maxv - df.minv
        df['bestnorm'] = (1.0 * df.best - df.minv) / diffv
        secbestnorm = (1.0 * df.secbest - df.minv) / diffv
        df['secbestnorm'] = secbestnorm.fillna(secbestnorm.min())
        df['diffnorm'] = df.bestnorm - df.secbestnorm

    for df in [dfs['c'], dfs['d']]:
        diffv_1 = df.maxv1 - df.minv1
        df['bestnorm1'] = (1.0 * df.best1 - df.minv1) / diffv_1
        secbestnorm_1 = (1.0 * df.secbest1 - df.minv1) / diffv_1
        df['secbestnorm1'] = secbestnorm_1.fillna(secbestnorm_1.min())
        df['diffnorm1'] = df['bestnorm1'] - df['secbestnorm1']
    
        diffv_2 = df.maxv2 - df.minv2
        df['bestnorm2'] = (1.0 * df.best2 - df.minv2) / diffv_2
        secbestnorm_2 = (1.0 * df.secbest2 - df.minv2) / diffv_2
        df['secbestnorm2'] = secbestnorm_2.fillna(secbestnorm_2.min())
        df['diffnorm2'] = df['bestnorm2'] - df['secbestnorm2']
    
        diff_df = diffv_1 + diffv_2
        df['bestnormconc'] = (1.0 * df.best1 + df.best2 - df.minv1 - df.minv2) / diff_df
        secbest_df = (1.0 * df.secbest1 + df.secbest2 - df.minv1 - df.minv2) / diff_df
        df['secbestnormconc'] = secbest_df.fillna(secbest_df.min())
        df['diffnormconc'] = df['bestnormconc'] - df['secbestnormconc']

    return dfs
```

```python
import math

def pcor_to_mapq(p):
    return -10.0 * math.log10(1.0 - p) if p < 1.0 else float('inf')

def pcor_to_mapq_iter(p):
    return map(pcor_to_mapq, p)

def mapq_to_pcor(p):
    return (1.0 - (10.0 ** (-0.1 * p))) if p < float('inf') else 1.0

def mapq_to_pcor_iter(p):
    return map(mapq_to_pcor, p)

def round_pcor(pcor):
    assert not any(map(lambda x: x < 0.0 or x > 1.0, pcor))
    return mapq_to_pcor_iter(map(round, (pcor_to_mapq_iter(pcor))))

pcor_orig_test = mapq_to_pcor_iter(mapq_orig_test)
y_train = map(lambda x: x == 1, train.correct)
y_test = map(lambda x: x == 1, test.correct)
test_data_has_correct = test['correct'].count() == len(test['correct'])
```

```python
from collections import defaultdict

def _tally_cor_per(level, cor):
    tally = defaultdict(lambda: [0, 0])
    for p, c in zip(level, cor):
        c = 0 if c else 1
        tally[p][c] += 1
    return tally

def _drop_rate_cum_sum(pcor, cor):
    tally = _tally_cor_per(pcor, cor)
    cumsum, cumsums = 0, []
    for p, tup in sorted(tally.iteritems(), reverse=True):
        ncor, nincor = tup
        for i in xrange(ncor + nincor):
            cumsum += (float(nincor) / (ncor + nincor))
            cumsums.append(cumsum)
    return cumsums
```

```python
import matplotlib.pyplot as plt
import numpy as np

def _plot_drop_rate(pcor, cor, pcor2=None, log2ize=False):
    """  """
    cumsum = _drop_rate_cum_sum(pcor, cor)
    cumsum2 = None if pcor2 is None else _drop_rate_cum_sum(pcor2, cor)
    ln = len(pcor)
    x_log = np.linspace(1.0/ln, (ln - 1.0) / ln, ln - 1.0)
    if log2ize:
        x_log = -np.log2(x_log[::-1])
    maxx = math.ceil(np.log2(ln))
    fig = plt.figure(figsize=(10, 4))
    axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
    axes.plot(x_log, cumsum[:-1], color='r', label='Model')
    if cumsum2 is not None:
        axes.plot(x_log, cumsum2[:-1], color='k', label='Original')
    axes.set_xlabel('MAPQ drop rate')
    if log2ize:
        axes.set_xticklabels(map(lambda x: 2 ** x, np.linspace(-1, -maxx, maxx)), rotation=90)
    axes.set_ylabel('Cumulative # incorrect alignments')
    axes.set_title('Drop rate comparison on -log2 scale')
    axes.legend(loc=2)
    axes.grid(True)
    return fig
```

```python
def ranking_error(pcor, cor, rounded=False):
    assert len(pcor) == len(cor)
    if rounded:
        pcor = round_pcor(pcor)
    tally = _tally_cor_per(pcor, cor)
    err, sofar = 0, 0
    # from low-confidence to high confidence
    for p in sorted(tally.iterkeys()):
        ncor, nincor = tally[p]
        ntot = ncor + nincor
        assert ntot > 0
        if nincor > 0:
            # spread error over this grouping of tied pcors
            frac = float(nincor) / ntot
            assert frac <= 1.0
            err += frac * sum(xrange(sofar, sofar + ntot))
        sofar += ntot
    return err
```

```python
from scipy.interpolate import splrep, splev
from sklearn.metrics import mean_squared_error
from scipy.optimize import minimize_scalar
from scipy.optimize import brute

def smoothfit_error(pcor, cor, rounded=False, s=None, plot=False, log2ize=False, use_brute=True):
    """ Calculate the smooth-fit error.  Possibly also plot the smooth
        fit. """
    assert len(pcor) == len(cor)
    if rounded:
        pcor = round_pcor(pcor)
    tally = _tally_cor_per(pcor, cor)
    # create parallel pcor and cor lists
    sorted_pcor, sorted_cor = [], []
    for p in sorted(tally.iterkeys(), reverse=True):
        ncor, nincor = tally[p]
        ntot = ncor + nincor
        sorted_pcor.extend([p] * ntot)
        # use fractional correctness so results are deterministic
        sorted_cor.extend([float(ncor)/ntot] * ntot)
    m = len(sorted_cor)
    assert m == len(sorted_pcor) == len(pcor)
    
    if s is None:
        def _error(s):
            y = splev(xrange(m), splrep(xrange(m), sorted_cor, s=s))
            return mean_squared_error(sorted_pcor, y)
        
        if use_brute:
            s = brute(_error, [(50, 200, 25)])[0]
        else:
            res = minimize_scalar(_error, method='bounded', bounds=(50, 200), options={'maxiter':30})
            s = res.x
    
    y = splev(xrange(m), splrep(xrange(m), sorted_cor, s=s))
    
    if plot:
        ln = len(pcor)
        x = np.linspace(1.0/ln, (ln - 1.0) / ln, ln - 1.0)
        if log2ize:
            x = -np.log2(x[::-1])
        maxx = math.ceil(np.log2(ln))
        fig = plt.figure(figsize=(12, 4))
        axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
        axes.plot(x, sorted_pcor[:-1], color='r', label='Pcor')
        axes.plot(x, y[:-1], color='k', label='Smoothed cor')
        axes.set_xlabel('Index')
        if log2ize:
            axes.set_xticklabels(map(lambda x: 2 ** x, np.linspace(-1, -maxx, maxx)), rotation=90)
        axes.set_ylabel('Probability')
        axes.set_title('Smooth-fit plot')
        axes.legend(loc=3)
        axes.grid(True)

    # smaller is better
    return mean_squared_error(sorted_pcor, y), y, s
```

```python
import random

class MapqPredictions:
    
    def __init__(self):
        self.pcor, self.mapq_orig, self.correct = [], [], None
    
    def add_pcors(self, pcor, mapq_orig, correct=None):
        self.pcor.extend(pcor)
        self.mapq_orig.extend(mapq_orig)
        if correct is not None:
            if self.correct is None:
                self.correct = []
            self.correct.extend(correct)
    
    def finalize(self):
        pcor, mapq = self.pcor, pcor_to_mapq_iter(self.pcor)
        self.mapq = mapq
        mapq_orig_test = self.mapq_orig
        pcor_orig_test = mapq_to_pcor_iter(mapq_orig_test)
        
        # calculate error measures and other measures
        
        if self.correct is not None:
            # ranking error
            self.rank_err_orig = ranking_error(pcor_orig_test, self.correct)
            self.rank_err = ranking_error(pcor, self.correct) / self.rank_err_orig
            self.rank_err_round = ranking_error(pcor, self.correct, rounded=True) / self.rank_err_orig
            
            # smooth-fit error
            self.sfit_err, y, s = smoothfit_error(pcor, self.correct)
            self.sfit_err_orig, y_orig, s_orig = smoothfit_error(pcor_orig_test, self.correct)
            self.sfit_err = self.sfit_err / self.sfit_err_orig
            self.sfit_smoothness = s
        
        # summary statistics over pcors and mapqs
        self.pcor_avg, self.pcor_orig_avg = np.mean(pcor), np.mean(pcor_orig_test)
        self.mapq_avg, self.mapq_orig_avg = np.mean(mapq), np.mean(mapq_orig_test)
        self.pcor_std, self.pcor_orig_std = np.std(pcor), np.std(pcor_orig_test)
        self.mapq_std, self.mapq_orig_std = np.std(mapq), np.std(mapq_orig_test)
    
    def summary(self):
        return "rank_err=%0.5f,%0.5f rounded, sfit_err=%0.5f, mapq_avg=%0.3f, mapq_std=%0.3f" % \
            (self.rank_err, self.rank_err_round, self.sfit_err, self.mapq_avg, self.mapq_std)

class MapqFit:
    
    @staticmethod
    def _dfs_to_train_test_mapq(train, test, paired):
        if paired:
            # Extract relevant paired-end features from training data frame
            x_train = train[['bestnorm1', 'diffnorm1', 'diffnormconc']].values
            x_test = test[['bestnorm1', 'diffnorm1', 'diffnormconc']].values
            mapq_header = 'mapq1'
        else:
            # Extract relevant unpaired features from training data frame
            x_train = train[['bestnorm', 'diffnorm']].values
            x_test = test[['bestnorm', 'diffnorm']].values
            mapq_header = 'mapq'
        # Also extract original MAPQs predicted by the aligner
        mapq_orig_train, mapq_orig_test = train[mapq_header], test[mapq_header]
        y_train = map(lambda x: x == 1, train.correct)
        return x_train, x_test, mapq_orig_train, mapq_orig_test, y_train
    
    @staticmethod
    def _downsample(x_train, mapq_orig_train, y_train, sample_fraction):
        n_training_samples = x_train.shape[0]
        if sample_fraction < 1.0:
            sample_indexes = random.sample(xrange(n_training_samples), int(round(n_training_samples * sample_fraction)))
            x_train = x_train[sample_indexes, ]
            mapq_orig_train = mapq_orig_train[sample_indexes]
            y_train = [y_train[i] for i in sample_indexes]
        return x_train, mapq_orig_train, y_train
    
    def summary(self):
        return self.pred_overall.summary()
    
    def __init__(self, dr, model, sample_fraction=1.0):
        self.train_dfs = read_dataset(os.path.join(dr, 'training'))
        self.test_dfs = read_dataset(os.path.join(dr, 'test'))
        self.model = model
        datasets = zip('udcb',
                       ['Unpaired', 'Discordant', 'Concordant', 'Bad-end'],
                       [False, True, True, False])
        self.pred_per_dataset = {}
        self.pred_overall = MapqPredictions()
        for dataset_shortname, dataset_name, paired in datasets:
            train, test = train_dfs[dataset_shortname], test_dfs[dataset_shortname]
            y_test = map(lambda x: x == 1, test.correct)
            if test.shape[0] == 0:
                continue  # No test data of this type
            assert train.shape[0] > 0  # Must be at least some training data of this type
            self.pred_per_dataset[dataset_shortname] = MapqPredictions()
            # extract features
            x_train, x_test, mapq_orig_train, mapq_orig_test, y_train = self._dfs_to_train_test_mapq(train, test, paired)
            # optionally downsample
            x_train, mapq_orig_train, y_train = self._downsample(x_train, mapq_orig_train, y_train, sample_fraction)
            model.fit(x_train, y_train)  # fit model
            pcor_test = model.predict(x_test)  # make predictions
            if any(np.isinf(pcor_test)):
                max_noninf_pcor_test = max(filter(lambda x: not np.isinf(x), pcor_test))
                pcor_test = [max_noninf_pcor_test if np.isinf(p) else p for p in pcor_test]
            for pred in [self.pred_per_dataset[dataset_shortname], self.pred_overall]:
                pred.add_pcors(pcor_test, mapq_orig_test, y_test)
        self.pred_overall.finalize()
        for pred in self.pred_per_dataset.itervalues():
            pred.finalize()
```

```python
from multiprocessing import Pool
from functools import partial

def fit(dr, model, sample_fraction):
    return MapqFit(dr, model, sample_fraction)

def downsampled_series(dr, model):
    fractions = [1.0, 0.50, 0.30, 0.20, 0.10, 0.05, 0.025, 0.02]
    fit_partial = partial(fit, dr, model)
    p = Pool(len(fractions))
    preds = p.map(fit_partial, fractions)
    perf_dict = {'fraction':fractions,
                 'rank_err':[pred.pred_overall.rank_err for pred in preds],
                 'rank_err_round':[pred.pred_overall.rank_err_round for pred in preds],
                 'sfit_err':[pred.pred_overall.sfit_err for pred in preds],
                 'smoothness':[pred.pred_overall.sfit_smoothness for pred in preds],
                 'pcor_avg':[pred.pred_overall.pcor_avg for pred in preds],
                 'mapq_avg':[pred.pred_overall.mapq_avg for pred in preds],
                 'pcor_std':[pred.pred_overall.pcor_std for pred in preds],
                 'mapq_std':[pred.pred_overall.mapq_std for pred in preds]}
    return pandas.DataFrame.from_dict(perf_dict)
```

```python
simple_dir = os.path.join(os.environ['TS_HOME'], 'data', 'simulated_reads', 'simple')

unpaired_dir = os.path.join(simple_dir, 'r0_bt2s_mason_ill_100_100k.ts')
model_35_6 = RandomForestRegressor(n_estimators=35, max_depth=6, random_state=33, max_features='auto')
series_unpaired_35_6 = downsampled_series(unpaired_dir, model_35_6)

model_30_5 = RandomForestRegressor(n_estimators=30, max_depth=5, random_state=77, max_features='auto')
series_unpaired_30_5 = downsampled_series(unpaired_dir, model_30_5)

model_25_4 = RandomForestRegressor(n_estimators=25, max_depth=4, random_state=25, max_features='auto')
series_unpaired_25_4 = downsampled_series(unpaired_dir, model_25_4)
```

```python
def plot_downsampling_series(seriess, labs=None, colors=None):
    fig = plt.figure(figsize=(14, 8))

    if labs is None:
        labs = [str(i+1) for i in xrange(len(seriess))]
    if colors is None:
        colors = ['r', 'b', 'g', 'o'][:len(seriess)]
        assert len(colors) == len(seriess)
    
    ax1 = fig.add_subplot(2, 1, 1)
    for series, lab, color in zip(seriess, labs, colors):
        ax1.plot(series.fraction, series.rank_err_round, color=color, label=lab)
    ax1.set_xlabel('Fraction')
    ax1.set_ylabel('Round rank err')
    ax1.set_ylim([0.8, 1.2])
    ax1.legend(loc=1)
    ax1.grid(True)
    
    ax2 = fig.add_subplot(2, 1, 2)
    for series, lab, color in zip(seriess, labs, colors):
        ax2.plot(series.fraction, series.sfit_err, color=color, label=lab)
    ax2.set_xlabel('Fraction')
    ax2.set_ylabel('Smooth-fit error')
    ax2.set_ylim([0, 1.])
    ax2.legend(loc=1)
    ax2.grid(True)
```

### Bucket plot with MAPQ-defined buckets

```python
def bucket_error_plot(mapq_lists, labs, colors, cor, title=None):
    """ Plot predicted MAPQ versus actual MAPQ, both rounded to
        nearest integer.  Omit """
    observed_mapq_lists = []
    actual_mapq_lists = []
    mx = 0
    for mapq_list in mapq_lists:
        # round MAPQs off to integers
        mapq = map(round, mapq_list)
        # stratify by mapq & count correct/incorrect
        tally = _tally_cor_per(mapq, cor)
        observed_mapq_lists.append([])
        actual_mapq_lists.append([])
        for p, tup in sorted(tally.iteritems()):
            ncor, nincor = tup
            if nincor > 0:  # ignore buckets where all are correct
                ntot = ncor + nincor
                observed_mapq_lists[-1].append(p)
                actual_mapq = pcor_to_mapq(float(ncor) / ntot)
                actual_mapq_lists[-1].append(actual_mapq)
                mx = max(max(mx, p), actual_mapq)
    fig = plt.figure(figsize=(8, 8))
    axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
    axes.set_xlabel('Reported MAPQ')
    axes.set_ylabel('Actual MAPQ')
    if title is not None:
        axes.set_title(title)
    for o, a, l, c in zip(observed_mapq_lists, actual_mapq_lists, labs, colors):
        axes.scatter(o, a, label=l, color=c)
    axes.plot([0, mx], [0, mx], color='r')
    axes.set_xlim((0, mx))
    axes.set_ylim((0, mx))
    axes.legend(loc=2)
    axes.grid(True)
```

### Bucket plot with quantile-defined buckets

```python
import math

def quantile_error_plot(mapq_lists, labs, colors, cor, title=None, quantiles=10):
    # One problem is how to determine the quantiles when there are multiple methods being compared
    estimated_mapq_lists = []
    actual_mapq_lists = []
    mx = 0
    for mapq_list in mapq_lists:
        # stratify by mapq & count correct/incorrect
        n = len(mapq_list)
        assert n >= quantiles
        sorted_order = sorted(range(n), key=lambda x: mapq_list[x])
        mapq_sorted = [mapq_list[x] for x in sorted_order]
        pcor_sorted = mapq_to_pcor_iter(mapq_sorted)
        cor_sorted = [cor[x] for x in sorted_order]
        partition_starts = [int(round(i*n/quantiles)) for i in xrange(quantiles+1)]
        estimated_mapq_lists.append([])
        actual_mapq_lists.append([])
        for i in xrange(quantiles):
            st, en = partition_starts[i], partition_starts[i+1]
            ncor = sum(cor_sorted[st:en])
            assert ncor < (en-st)
            ncor_est = sum(pcor_sorted[st:en])
            estimated_mapq_lists[-1].append(pcor_to_mapq(float(ncor_est)/(en-st)))
            actual_mapq_lists[-1].append(pcor_to_mapq(float(ncor)/(en-st)))
        mx = max(mx, max(max(estimated_mapq_lists[-1]), max(actual_mapq_lists[-1])))
    assert not math.isinf(mx)
    fig = plt.figure(figsize=(8, 8))
    axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
    axes.set_xlabel('Reported MAPQ')
    axes.set_ylabel('Actual MAPQ')
    if title is not None:
        axes.set_title(title)
    for o, a, l, c in zip(estimated_mapq_lists, actual_mapq_lists, labs, colors):
        assert len(o) == len(a)
        axes.scatter(o, a, label=l, color=c, alpha=0.3, s=60)
    axes.plot([0, mx], [0, mx], color='r')
    axes.set_xlim((-.02*mx, 1.02*mx))
    axes.set_ylim((-.02*mx, 1.02*mx))
    axes.legend(loc=2)
    axes.grid(True)
```

### Visualizing the model fit

```python
import matplotlib.pyplot as plt
from matplotlib.colors import BoundaryNorm
from matplotlib.ticker import MaxNLocator
import numpy as np

def plot_fit(model, x_lim = (0.0, 1.0), y_lim = (0.0, 1.0)):
    # make these smaller to increase the resolution
    dx, dy = 0.02, 0.02

    # generate 2 2d grids for the x & y bounds
    y, x = np.mgrid[slice(0, y_lim + dy, dy),
                    slice(0, x_lim + dx, dx)]

    z = model.predict()

    # x and y are bounds, so z should be the value *inside* those bounds.
    # Therefore, remove the last value from the z array.
    z = z[:-1, :-1]
    levels = MaxNLocator(nbins=15).tick_values(z.min(), z.max())

    # pick the desired colormap, sensible levels, and define a normalization
    # instance which takes data values and translates those into levels.
    cmap = plt.get_cmap('PiYG')
    norm = BoundaryNorm(levels, ncolors=cmap.N, clip=True)

    plt.subplot(2, 1, 1)
    im = plt.pcolormesh(x, y, z, cmap=cmap, norm=norm)
    plt.colorbar()
    # set the limits of the plot to the limits of the data
    plt.axis([x.min(), x.max(), y.min(), y.max()])
    plt.title('pcolormesh with levels')
```
